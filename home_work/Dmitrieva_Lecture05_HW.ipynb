# **2. Построить бинарный классификатор для каждого семейства(Family) на основе LogisticRegression c $l_1$ и $l_2$ регуляризацими. Это соответствует целевым исходам (Конкретное семейство, Всё остальное). Подобрать гиперпараметр $C$ с помощью кросс-валидации, используя метрику ROC AUC.**

**Подготовка данных**

Целевая переменная (y):  1 — конкретное семейство, 0 — остальные.

Признаки (X): Нормализованные MFCC-данные (df_normalized).

Важно: Убедимся, что MFCCs_1 (все значения = 1.0) не является константным признаком (если да — удалим).

--------------------------------------------
**Константный признак** - это признак, у которого все значения в датасете одинаковы (например, MFCCs_1 = 1.0 для всех строк)

Если признак не меняется, он не несёт информации для классификации/регрессии. Модель не сможет использовать его для разделения классов.

Проблемы при регуляризации:

*   Для L1-регуляризации (лассо) коэффициент такого признака станет 0, но это пустая трата вычислительных ресурсов.

*   Для L2-регуляризации (ридж) признак добавит шум в штраф функции потерь.

```python
df2 = df_normalized
```

```python
from sklearn.preprocessing import LabelEncoder #LabelEncoder — это инструмент, который преобразует категориальные (текстовые) метки в числовые значения.

# Удаление константного признака (если MFCCs_1 = 1.0 для всех строк)
if df2['MFCCs_1'].nunique() == 1:
    df2 = df2.drop(columns=['MFCCs_1'])

# Выделение признаков и меток семейств
X = df2.filter(regex='MFCCs_')  # Все колонки MFCC
                                #regex= — для поиска столбцов по регулярному выражению
families = df2['Family'].unique()  # Неповторяющиеся названия семейства
```

**Обучение бинарных классификаторов для каждого семейства**

Для каждого семейства:



*   Создаем бинарную метку: 1 для текущего семейства, 0 для остальных.

*   Разделяем данные на train/test.

*   Подбираем гиперпараметр C для L1 и L2 регуляризации через кросс-валидацию (GridSearchCV).

*   Оцениваем качество по ROC AUC.









```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import roc_auc_score #roc_auc_score — метрика ROC-AUC для оценки качества модели.

# Параметры для GridSearchCV
param_grid = {                            # Чем меньше C, тем сильнее штраф за большие веса (модель проще). Чем больше C, тем слабее регуляризация (модель сложнее, риск переобучения).
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Опорные значения для подбора. C=0.1 как оптимальное.
    'penalty': ['l1', 'l2'],              # penalty - штраф, 'l1' (лассо) — обнуляет некоторые веса (отбор признаков), 'l2' (ридж) — уменьшает веса равномерно.
    'solver': ['liblinear']               # liblinear поддерживает оба типа регуляризации L1 и L2
}

results = {}
```

```python
for family in families:
    # Бинарная метка: 1 для текущего семейства, 0 для остальных
    y = (df2['Family'] == family).astype(int)

    # Разделение на train/test
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Подбор гиперпараметров
    clf = GridSearchCV(
        LogisticRegression(max_iter=1000), #По умолчанию в sklearn max_iter=100. Этого может не хватить, особенно если данные высокоразмерные (много признаков, например, 22 MFCC)
        param_grid,
        cv=5,  # 5-fold кросс-валидация
        scoring='roc_auc'
    )
    clf.fit(X_train, y_train)

    # Предсказание на тесте
    y_pred_proba = clf.predict_proba(X_test)[:, 1] # вероятность класса 1 ("текущее семейство").
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # Сохранение результатов
    results[family] = {
        'best_C': clf.best_params_['C'],
        'best_penalty': clf.best_params_['penalty'],
        'roc_auc': roc_auc,
        'model': clf.best_estimator_
    }
```

```python
# Вывод результатов
for family, res in results.items():
    print(f"Семейство: {family}")
    print(f"  Лучший C: {res['best_C']}, Регуляризация: {res['best_penalty']}")
    print(f"  ROC-AUC: {res['roc_auc']:.3f}\n")
```

# **3. Добавить полиномиальные признаки до второго порядка. Построить бинарный классификатор для каждого семейства на основе LogisticRegression c $l_1$ и $l_2$ регуляризацими. Подобрать гиперпараметр $C$ с помощью кросс-валидации, используя метрику ROC AUC.**

**Добавление полиномиальных признаков**

**Полиномиальные признаки** — это искусственно созданные признаки, которые включают:

* Исходные признаки (например, MFCC-коэффициенты).

* Их степени (квадраты, кубы и т.д.).

* Взаимодействия между признаками (попарные произведения).
----------------------------------------------
Зачем они нужны?
* Улавливают нелинейные зависимости между признаками и целевой переменной.

* Улучшают качество модели, если связь между признаками и целевой переменной сложная (например, квадратичная).

* Полезны для линейных моделей (например, LogisticRegression), которые сами по себе не могут работать с нелинейностями.

```python
from sklearn.preprocessing import PolynomialFeatures

# Создаём полиномиальные признаки до 2-го порядка (модель учитывает не только исходные признаки (MFCCs_1, MFCCs_2, ...), но и Квадраты признаков (MFCCs_1², MFCCs_2², ...), попарные взаимодействия (MFCCs_1 * MFCCs_2, MFCCs_1 * MFCCs_3, ...).)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)  # X — исходные MFCC-признаки

# Проверяем размерность
print(f"Исходное число признаков: {X.shape[1]}")
print(f"После добавления полиномиальных: {X_poly.shape[1]}")
```

**Обучение бинарных классификаторов для каждого семейства**


```python
results_poly = {}

for family in families:
    # Бинарная метка: 1 для текущего семейства, 0 для остальных
    y = (df2['Family'] == family).astype(int)

    # Разделение на train/test
    X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.3, random_state=42)

    # Подбор гиперпараметров
    clf = GridSearchCV(
        LogisticRegression(max_iter=1000, class_weight='balanced'),  # Учёт дисбаланса
        param_grid,
        cv=5,
        scoring='roc_auc'
    )
    clf.fit(X_train, y_train)

    # Предсказание на тесте
    y_pred_proba = clf.predict_proba(X_test)[:, 1]
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # Сохранение результатов
    results_poly[family] = {
        'best_C': clf.best_params_['C'],
        'best_penalty': clf.best_params_['penalty'],
        'roc_auc': roc_auc,
        'model': clf.best_estimator_
    }

# Вывод результатов
for family, res in results_poly.items():
    print(f"Семейство: {family}")
    print(f"  Лучший C: {res['best_C']}, Регуляризация: {res['best_penalty']}")
    print(f"  ROC-AUC: {res['roc_auc']:.3f}\n")
```
